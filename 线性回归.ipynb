```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("Salary_Data.xls")
```

```python
x = data["YearsExperience"]
y = data["Salary"]
```

```python
plt.grid(linestyle=":")
plt.scatter(x, y, label="Sample Points")
w0, w1, lrate = 1, 1, 0.001
times = 100
"""
loss = 1/2 * sumi(1,n)(w0 + w1xi - yi)^2
loss关于w0的偏导：sum[i](1,n)(w0 + w1xi - yi)
loss关于w1的偏导：sum[i](1,n)(w0 + w1xi - yi) * xi
"""

for i in range(times):
    d0 = (w0 + w1 * x - y).sum()
    d1 = ((w0 + w1 * x - y) * x).sum()
    
    """
    更新w0, w1
    """
    w0 = w0 - lrate * d0
    w1 = w1 - lrate * d1
    
pred_y = w0 + w1 * x    
plt.plot(x, pred_y,color="red",label="Regression Line")
plt.legend()
```

```python
plt.grid(linestyle=":")
plt.scatter(x, y, label="Sample Points")
w0, w1, lrate = 1, 1, 0.001
times = 100
import sklearn.linear_model as lm
model = lm.LinearRegression()
train_x = pd.DataFrame(x)
train_y = y
"""
在model.fit(x,y)中，x必须是二维数组,注意train_x是什么
"""
model.fit(train_x,train_y)
pred_y = model.predict(train_x)
plt.plot(x,pred_y,color="red")
```
